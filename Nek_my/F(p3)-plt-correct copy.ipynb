```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
u, a30 = sp.symbols('u a30')
b = 1.77
I3_mod = 2*a30 * (3 * (1 - a30**2) * a30 + 2 * sp.cos(u) * (1 - a30**2)**(3/2) + b*a30)
a3val=0.5**0.5
I3_mod0 = I3_mod.subs({a30: a3val, u: 0}) 

print(I3_mod0)
```

```python
y, p3 = sp.symbols('y p3')
a3val=0.5**0.5
p3_values = np.arange(0, 1.1, 0.001)

#print(b_values)
y = 1

F = -13*p3**4 + p3**3 * (30 + 6*b) - p3**2 * (21 + 6*b + b**2 + I3_mod0*3) + p3*(4 + I3_mod0*b + I3_mod0*3) - I3_mod0**2/4
print(F)
#print(F.subs(p3,a3val**2).expand())
```

```python
# Список для хранения результатов
results = []

# Решаем уравнение для различных значений b
for elem in p3_values:
    f = F.subs({p3: elem}) 
    print(f)
    if f >= 0:
        results.append((elem, f))

print(len(results))
print("results:")
print(*results, sep='\n')
```

```python
x, y = zip(*results)  # Распаковываем кортежи в два списка

plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o', linestyle='None', color='b', label='Data Points')

plt.title('График значений')
plt.xlabel('значение p3')
plt.ylabel('значение F(p3)')
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```
