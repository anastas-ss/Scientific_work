```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Определяем символы
a3, b = sp.symbols('a3 b')

# Определяем уравнение
equation1 = 12*a3 - 24*a3**3 + 4*(3*(1-a3**2)**(1/2)*a3**2 + (1-a3**2)**(3/2)) + 4*b*a3

# Список для хранения результатов
b_values = np.arange(-1, 1.1, 0.1)
print(b_values)
```

```python
sol_1 = []
for b_val in b_values:
    solution1 = sp.solve(equation1.subs(b, b_val), a3)
    real_solutions1 = [sol.evalf() for sol in solution1 if sol.is_real]
    for elem in real_solutions1:
        sol_1.append((b_val, elem))
for elem in sol_1:
    print(elem)
```

```python
# Список для хранения результатов
results = []

# Решаем уравнение для различных значений b
for elem in sol_1:
    b_val = elem[0]
    real_solutions1 = elem[1]
    if real_solutions1:
        a3_val = real_solutions1  # Берем первое реальное решение
        # Подставляем в функцию f
        f = 6*a3_val**2 - 6*a3_val**4 + 4*a3_val*(1-a3_val**2)**(3/2) + 2*b_val*a3_val**2
        results.append((b_val, f))
f_a3_0 = [6*0**2 - 6*0**4 + 4*0*(1-0**2)**(3/2) + 2*b_val*0**2 for b_val in b_values]
f_a3_1 = [6*1**2 - 6*1**4 + 4*1*(1-1**2)**(3/2) + 2*b_val*1**2 for b_val in b_values]
#results.append((np.float64(0), np.float64(0))) #все значения = 0
results.append((np.float64(1), np.float64(min(f_a3_1))))
results.append((np.float64(1), np.float64(max(f_a3_1))))
print(len(results))
print("results:")
print(*sorted(results), sep="\n")
```

```python
x, y = zip(*results)  # Распаковываем кортежи в два списка

plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Data Points')

plt.scatter(x[-2:], y[-2:], color='r', s=100, label='max и min при a3=1') 

plt.title('График значений')
plt.xlabel('значение b')
plt.ylabel('значение f')
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```
